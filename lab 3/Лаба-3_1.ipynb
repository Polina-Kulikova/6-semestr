```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.sample(10)
```

#### Имена сортов

```python
print(iris.target_names)
```

#### Добавим столбец target

```python
df['target']=iris.target
df
```

#### Построим графики зависимостей

```python
plt.xlabel('sepal width (cm)')
plt.ylabel('sepal length (cm)')
plt.title('График зависисмости sepal length - sepal width')


plt.scatter(df[df['target']==0]['sepal width (cm)'], df[df['target']==0]['sepal length (cm)'], color='g', label = 'setosa')
plt.scatter(df[df['target']==1]['sepal width (cm)'], df[df['target']==1]['sepal length (cm)'], color='b', label = 'versicolor')
plt.scatter(df[df['target']==2]['sepal width (cm)'], df[df['target']==2]['sepal length (cm)'], color='m', label = 'virginica')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.xlabel('sepal width (cm)')
plt.ylabel('sepal length (cm)')
plt.title('График зависисмости petal length - petal width')

plt.scatter(df[df['target']==0]['petal width (cm)'], df[df['target']==0]['petal length (cm)'], color='g', label = 'setosa')
plt.scatter(df[df['target']==1]['petal width (cm)'], df[df['target']==1]['petal length (cm)'], color='b', label = 'versicolor')
plt.scatter(df[df['target']==2]['petal width (cm)'], df[df['target']==2]['petal length (cm)'], color='m', label = 'virginica')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue='target', palette="tab10")
```

#### Датасет, где есть setosa и versicolor

```python
df_1 = df.drop(df[df['target']==2].index)
df_1
```

#### Датасет, где versicolor и virginica

```python
df_2=df.drop(df[df['target']==0].index)
df_2.info()
```

```python
df_1.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_1 = df_1
y_1 = df_1['target']
```

```python
X_1 = X_1.drop(['target'], axis=1)
X_1
```

```python
y_1
```

#### Разбиваем df_1 (датасет, где есть setosa и versicolor) на обучающую и тестовые выборки

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X_1, y_1, test_size=0.2, random_state=42)
```

```python
clf = LogisticRegression(random_state=0)
clf.fit(X1_train, y1_train)
predict_1 = clf.predict(X1_test)
```

#### Сравниваем реальные данные и предсказанные

```python
list(zip(y1_test, predict_1))
```

#### Точность модели

```python
accuracy_score(y1_test, predict_1)
```
